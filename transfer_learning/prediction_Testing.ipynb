```
from google.colab import drive
drive.mount('/content/drive/')
%cd /content/drive/My\ Drive/Colab\ Notebooks\

#=============================================================================
from tensorflow.keras.models import load_model
import scipy.io as sio
import numpy as np

#=============================================================================
print('Loading data...')
ExType = 'H18'
DataPath = './wind_data/Norm_Test_data_Lag10_'+ExType+'.mat'   
test_data = sio.loadmat(DataPath)['XTest']#[...,np.newaxis]
test_labels = sio.loadmat(DataPath)['YTest']
test_labels = np.squeeze(test_labels)

print('Wind Test data dim.   (samples x lag x parameters)',test_data.shape)
print('Wind Test labels dim. (samples x lag x parameters)',test_labels.shape)


# data resizing
resizeTo = 32
from tensorflow.keras.preprocessing.image import array_to_img
test_images = np.zeros((test_data.shape[0],resizeTo,resizeTo))
for i, im in enumerate(test_data):
  a = array_to_img(test_data[i,:, :, np.newaxis])
  test_images[i,...] = np.array(a.resize((resizeTo,resizeTo)))


# duplicate the data three times to look like rgb image
test_data = np.repeat(test_images[:, :, :, np.newaxis], 3, axis=3)
print('Wind Test images dim.   (samples x lag x parameters x channels)',test_data.shape)
#=============================================================================
```

```
# ========================================================= H3, H6, H12, H18
# filename = 'vgg16_tuning' 
# filename = 'vgg19_tuning' 
# filename = 'ResNet50_tuning'
# filename = 'ResNet101_tuning'
# filename = 'ResNet152_tuning'
# filename = 'DenseNet121_tuning' 
# filename = 'DenseNet169_tuning' 
filename = 'DenseNet201_tuning' 
#========================================================= H1
# filename = 'vgg16_tuning' 
# filename = 'vgg19_tuning' 
# filename = 'mobilenetV2_tuning'
# filename = 'Xception_tuning'
# filename = 'ResNet50_tuning'
# filename = 'ResNet101_tuning'
# filename = 'ResNet152_tuning'
# filename = 'ResNet50V2_tuning' 
# filename = 'ResNet101V2_tuning' 
# filename = 'ResNet152V2_tuning' 
# filename = 'InceptionV3_tuning' xxx
# filename = 'InceptionResNetV2_tuning' xxx
# filename = 'MobileNet_tuning' 
# filename = 'DenseNet121_tuning' 
# filename = 'DenseNet169_tuning' 
# filename = 'DenseNet201_tuning' 
# filename = 'NASNetMobile_tuning' 
# filename = 'NASNetLarge_tuning' 
# filename = 'EfficientNetB0_tuning' 
# filename = 'EfficientNetB1_tuning'
# filename = 'EfficientNetB2_tuning'
# filename = 'EfficientNetB3_tuning'
# filename = 'EfficientNetB4_tuning'
# filename = 'EfficientNetB5_tuning'
# filename = 'EfficientNetB6_tuning'
# filename = 'EfficientNetB7_tuning'


saveTo = './models/'+ExType
modelName = saveTo + '/best_'+filename+'.hdf5'
model = load_model(modelName)
predictions = model.predict(test_data,verbose=1).squeeze()
#=============================================================================
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt

# MAPE = np.mean(100*2.0 * np.abs(valid_labels - predictions) / (np.abs(valid_labels) + np.abs(predictions))).item()
MAPE = np.mean(np.abs(test_labels - predictions) / np.abs(test_labels)) * 100
print("MAE {:.4}, MSE {:.4f}, RMSE {:.4f}, MAPE {:.4f}".format(
    mean_absolute_error(test_labels, predictions),
    mean_squared_error(test_labels, predictions),
    sqrt(mean_squared_error(test_labels, predictions)),
    MAPE))
plt.scatter(test_labels,predictions)
plt.figure(figsize=(30,6)); 
plt.plot(test_labels, label='true')
plt.plot(predictions, label='pred')
plt.legend()
```
